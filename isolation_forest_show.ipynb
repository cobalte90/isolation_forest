```python
import numpy as np
import matplotlib.pyplot as plt
from models.isolation_forest import IsolationForest

np.random.rand()
normal_data = np.random.randn(1000, 2) * 0.5
outliers = np.random.uniform(low=-4, high=4, size=(100, 2))
X = np.vstack([normal_data, outliers])

model = IsolationForest(contamination=0.1)
model.fit(X)
labels = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm', s=20)
plt.title("Isolation Forest")
plt.colorbar(ticks=[-1, 1], label='Class: (1=normal, -1=ourliers)')
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import IsolationForest as sklearn_IF
model = sklearn_IF(contamination=0.1)
model.fit(X)
sklearn_pred = model.predict(X) * -1
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=sklearn_pred, cmap='coolwarm', s=20)
plt.title("Sklearn Isolation Forest")
plt.colorbar(ticks=[-1, 1], label='Class: (1=normal, -1=ourliers)')
plt.grid(True)
plt.show()
```
